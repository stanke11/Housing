```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Housing.csv')
```

```python
print(df.head())
```

```python
print(df.describe())
```

```python
print(df.info())
```

```python
# Scatter plot between "area" and "price"
sns.scatterplot(x='area', y='price', data=df)
```

```python
# Histogram of "price"
sns.histplot(data=df, x='price')
```

```python
# Box plot of "price" based on "furnishingstatus"
sns.boxplot(x='furnishingstatus', y='price', data=df)
```

```python
# Count plot of "furnishingstatus"
sns.countplot(x='furnishingstatus', data=df)
```

```python
# Bar plot of "furnishingstatus" and "price"
sns.barplot(x='furnishingstatus', y='price', data=df)
```

```python
# Heatmap of the correlation between "price," "area," "bedrooms," "bathrooms," and "stories"
sns.heatmap(df[['price', 'area', 'bedrooms', 'bathrooms', 'stories']].corr(), annot=True)
```

```python
# Pair plot of "price," "area," "bedrooms," "bathrooms," and "stories"
sns.pairplot(df[['price', 'area', 'bedrooms', 'bathrooms', 'stories']])
```
